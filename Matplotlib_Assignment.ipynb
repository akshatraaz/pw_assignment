```python
Q2. What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

Ans. A Scater plot is a diagram where each value in the data set is represented by a dot. The Matplotlib Module is the method for drawing 
 Scatter Plots. It needs two arrays of the same length, one for the values of the x- axis, and the one for the values of y- axis
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel('This is the Data of X')
plt.ylabel('This is the Data of Y')
plt.title("Title - Scatter Plot Graph")
```

```python
Q3. Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])


Ans. There are many reason to using subplot() function in Matplotlib like we can use subplot() to create multiple plot, custumizing plot layout, saving
 space and faceting etc.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure and axis object
fig, axs = plt.subplots(2, 2, figsize=(8, 6))

# Plot 1
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# Plot 2
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Plot 3
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Plot 4
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Adjust layout and spacing
fig.tight_layout()

# Show the plot
plt.show()
```

```python
Q5. What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

Ans. Box plots are used to show distribution of numeric data values, Especially when we want to compare between multiple groups. 
```

```python
import matplotlib.pyplot as plt
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
# Create a figure and axis object
fig, ax = plt.subplots()

# Create a boxplot
ax.boxplot([box1, box2])

# Add x-axis labels
ax.set_xticklabels(['Box 1', 'Box 2'])

# Add a title to the plot
ax.set_title('Box Plot of Box 1 and Box 2')

# Show the plot
plt.show()

```
